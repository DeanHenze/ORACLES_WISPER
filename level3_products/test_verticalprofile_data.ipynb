# Test plots for level 3 vertical profile files

If ```vertical_profiles.py``` runs succesfully and procudes the correct data files, the output from this notebook should be as in the pdf version (also within this directory).

````python
import numpy as np
import xarray as xr
import netCDF4 as nc # 1.3.1
import matplotlib.pyplot as plt

%matplotlib inline
````

### Load and print data

````python
prfs2016 = xr.load_dataset("wisper_oracles_verticalprofiles_2016.nc", decode_times=False)
print(prfs2016)
print("\n\n")
print(prfs2016['q'])
````

````python
prfs2017 = xr.load_dataset("wisper_oracles_verticalprofiles_2017.nc", decode_times=False)
print(prfs2017)
print("\n\n")
print(prfs2017['T'])
print(prfs2017['dD'])
````

````python
prfs2018 = xr.load_dataset("wisper_oracles_verticalprofiles_2018.nc", decode_times=False)
print(prfs2018)
print("\n\n")
print(prfs2018['P'])
print(prfs2018['d18O'])
````

### Basic check, all profiles for temperature, pressure, water, cwc, total water isotope ratios

**2016 sampling period**

````python
fig, axset = plt.subplots(3, 2, figsize=(10,12))
for pnum in prfs2016['profile']:

    prfsingle = prfs2016.sel(profile=pnum)
    axset[0,0].plot(prfsingle['T'], prfsingle['alt'])
    axset[0,1].plot(prfsingle['P'], prfsingle['alt'])
    axset[1,0].plot(prfsingle['q'], prfsingle['alt'])
    #axset[1,1].plot(prfsingle['cwc'], prfsingle['alt'])
    axset[2,0].plot(prfsingle['dD'], prfsingle['alt'])
    axset[2,1].plot(prfsingle['d18O'], prfsingle['alt'])
    
axset[1,0].set_ylabel('altitude, meters', fontsize=14)
plt_xlabels = ['T, K', 'P, hPa', 'q, q/kg', 'cwc, g/m3', 'dD, permil', 'd18O, permil']
for ax, lab in zip(axset.flatten(), plt_xlabels):
    ax.set_xlabel(lab, fontsize=14)
````

**2017 sampling period**

````python
fig, axset = plt.subplots(3, 2, figsize=(10,12))
for pnum in prfs2017['profile']:

    prfsingle = prfs2017.sel(profile=pnum)
    axset[0,0].plot(prfsingle['T'], prfsingle['alt'])
    axset[0,1].plot(prfsingle['P'], prfsingle['alt'])
    axset[1,0].plot(prfsingle['q'], prfsingle['alt'])
    axset[1,1].plot(prfsingle['cwc'], prfsingle['alt'])
    axset[2,0].plot(prfsingle['dD'], prfsingle['alt'])
    axset[2,1].plot(prfsingle['d18O'], prfsingle['alt'])
    
axset[1,0].set_ylabel('altitude, meters', fontsize=14)
plt_xlabels = ['T, K', 'P, hPa', 'q, q/kg', 'cwc, g/m3', 'dD, permil', 'd18O, permil']
for ax, lab in zip(axset.flatten(), plt_xlabels):
    ax.set_xlabel(lab, fontsize=14)
````
